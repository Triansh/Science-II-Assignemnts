```python
import matplotlib.pyplot as plt
import numpy as np

tot_plots = 6
iterations = 500
dp = [x for x in range(1,tot_plots + 1)]
dx = [0 for x in range(1,tot_plots + 1)]
k = .5
m = 5
X = np.zeros((tot_plots, iterations))
P = np.zeros((tot_plots, iterations))
H = np.zeros(tot_plots)
```

```python

for i in range(tot_plots):
    for j in range(iterations):
        X[i][j] = dx[i]
        P[i][j] = dp[i]
        dx[i] = dx[i] + dp[i]/m *.5
        dp[i] = dp[i] - k*dx[i] * .5
    H[i] = 0.5 * (((dx[i]**2) *k) + ((dp[i]**2) * (1/m)))
```

```python
def get_PvsX():
    plt.figure(figsize=(15, 10))
    for i in range(tot_plots):
        plt.plot(X[i], P[i], label = f"H={round(H[i], 6)}")
    plt.ylabel("Momentum")
    plt.xlabel("Displacement")
    plt.legend(loc = 'upper right')
    plt.show()

def mean_square():
    plt.figure(figsize=(20, 10))
    for i in range(tot_plots):
        plt.plot(X[i]**2, label = f"H={round(H[i], 6)}")
    plt.xlabel("Time")
    plt.ylabel("Mean square Displacement")
    plt.legend()
    plt.show()
    
```

```python
get_PvsX()
mean_square()
```
