```python
import matplotlib.pyplot as plt
import numpy as np

deltaX = 0.125
deltaTime = 0.001
D = 5
L = 6

X = np.arange(-L, L, deltaX)
P = np.zeros(len(X))
```

```python
def reset():
    global P
    P = np.array([(1 if x == 0 else 0) for x in X])


def update():
    global P
    z = np.zeros(P.shape)
    for i in range(1, len(P) - 1):
        z[i] = P[i] + (D * deltaTime / (deltaX ** 2)) * (P[i + 1] - 2 * P[i] + P[i - 1])
    P = z


def make_plot(n):
    reset()
    for i in range(n): update()
    plt.plot(X, P, label=f"n = {n}")
```

```python
plt.figure(figsize=(20, 10))
make_plot(10)
make_plot(100)
make_plot(1000)
make_plot(10000)
plt.legend(loc='upper right')
plt.show()
```
