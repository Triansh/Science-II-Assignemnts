```python
import matplotlib.pyplot as plt
import numpy as np

deltaX = 0.125
deltaY = 0.125
deltaTime = 0.001
Dx = [2, 1, 3]  # Change value of different Dx and Dy
Dy = [2, 3, 1]
n = [10, 100, 1000, 10000]  # change n here
L = 1

X = np.arange(-L, L, deltaX)
Y = np.arange(-L, L, deltaY)
M = np.zeros((len(X), len(Y)))
```

```python
def init():
    global M
    for i in range(len(M)):
        for j in range(len(M[i])):
            M[i][j] = 1 if (X[i] == 0 and Y[j] == 0) else 0


def update(difx, dify):
    global M
    Z = np.zeros(M.shape)

    for i in range(1, len(M) - 1):
        for j in range(1, len(M[i]) - 1):
            xx = difx / (deltaX ** 2) * (M[i + 1][j] - 2 * M[i][j] + M[i - 1][j])
            yy = dify / (deltaY ** 2) * (M[i][j + 1] - 2 * M[i][j] + M[i][j - 1])
            Z[i][j] = M[i][j] + deltaTime * (xx + yy)
    M = Z
```

```python

def make_plot(n):
    for i in range(3):
        init()
        fig, ax = plt.subplots(1, 1, figsize=(10, 10))
        plt.title(f"Dx={Dx[i]} Dy={Dy[i]} n={n}")
        for j in range(n): update(Dx[i], Dy[i])
        ax.contourf(X, Y, M, antialiased=False)
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        plt.show()


for num in n:
    make_plot(num)

```
