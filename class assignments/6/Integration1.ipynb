```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
def make_plot(title, xlab, ylab, plots):
    plt.figure(figsize=(10, 10))
    plt.title(title)
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    for plot in plots:
        plt.plot(plot['points'], label=plot['label'])
    plt.legend(loc='upper right')
    plt.show()


def get_deviation(l):
    return np.sqrt(np.var(np.array(l)))
```

```python
max_area = 1 * 3
simulation = []
iterations = 1000

for n in range(1, iterations):
    cnt = 0
    for i in range(n):
        x = random.random()
        y = random.random() * 3
        if y <= 3 * x ** 2:
            cnt += 1
    simulation.append((cnt / n) * max_area)

plots = [
    {'points': simulation, 'label': "Simulation"},
    {'points': [1] * iterations, 'label': "Actual"},
]

make_plot("Area vs N for I1", 'N', 'Area under graph', plots)

```

```python
def get_deviation_for_trials(N):
    max_area = 1 * 3
    simulation = []
    iterations = 100
    for t in range(iterations):
        cnt = 0
        for i in range(N):
            x = random.random()
            y = random.random() * 3
            if y <= 3 * x ** 2:
                cnt += 1
        simulation.append((cnt / N) * max_area)

    plots = [
        {'points': simulation, 'label': "Simulation"},
        {'points': [1] * iterations, 'label': "Actual"},
    ]
    make_plot(f"I1 vs Trials for N={N}", "Trials", "Integral", plots)
    print("The Standard Deviation is : ", get_deviation(simulation))


get_deviation_for_trials(20)
print()
get_deviation_for_trials(1000)
```
