```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
def make_plot(title, xlab, ylab, plots):
    plt.title(title)
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    for plot in plots:
        plt.plot(plot['points'], label=plot['label'])
    plt.legend(loc='upper right')
    plt.show()


def get_deviation(l):
    return np.sqrt(np.var(np.array(l)))


def get_3_random():
    return random.random(), random.random(), random.random()
```

```python
max_volume = 1
simulation = []
iterations = 1000

for n in range(1, iterations):
    cnt = 0
    for i in range(n):
        x, y, z = get_3_random()
        if z <= y * (x ** 2):
            cnt += 1
    simulation.append((cnt / n) * max_volume)

plots = [
    {'points': simulation, 'label': "Simulation"},
    {'points': [1 / 6] * iterations, 'label': "Actual"},
]

make_plot("Area vs N for I2", 'N', 'Area under graph', plots)

```

```python
def get_deviation_for_trials(N):
    max_volume = 1
    simulation = []
    iterations = 100
    for t in range(iterations):
        cnt = 0
        for i in range(N):
            x, y, z = get_3_random()
            if z <= y * (x ** 2):
                cnt += 1
        simulation.append((cnt / N) * max_volume)

    plots = [
        {'points': simulation, 'label': "Simulation"},
        {'points': [1 / 6] * iterations, 'label': "Actual"},
    ]
    make_plot(f"I2 vs Trials for N={N}", "Trials", "Integral", plots)
    print("The Standard Deviation is : ", get_deviation(simulation))


get_deviation_for_trials(20)
get_deviation_for_trials(1000)
```

```python
max_volume = 1 * 3
simulation = []
N = 1000
for points in range(1, N):
    area = []
    for t in range(100):
        cnt = 0
        for i in range(points):
            x, y, z = get_3_random()
            if z <= y * (x ** 2):
                cnt += 1
        area.append((cnt / points) * max_volume)
    simulation.append(get_deviation(area))

plots = [
    {'points': simulation, 'label': "Simulation"},
]
make_plot("Standard Deviation vs Number of trials", "N", "Standard Deivation", plots)
```
