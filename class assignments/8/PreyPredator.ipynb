```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.001


def solve_differential(a, b, c, k, r, n12):
    n1, n2 = n12
    dn1 = ((r * n1) - ((n1 ** 2) * r / k) - (a * n1 * n2)) * dt
    dn2 = ((b * n1 * n2) - (c * n2)) * dt
    return np.array([dn1, dn2])
```

```python
N0 = np.array([8, 8])
maxT = 30
```

```python
def prey_predator_model(a, b, c, k, r):
    time = np.arange(0, maxT, dt)
    iterations = len(time)
    pop = np.zeros((iterations, 2))
    pop[0] = N0

    for i in range(1, iterations):
        pop[i] = pop[i - 1] + solve_differential(a, b, c, k, r, pop[i - 1])
    plt.figure(figsize=(10, 6))
    plt.title(f'a={a} b={b} c={c} k={k} r={r}')
    plt.plot([x[0] for x in pop], label='Prey')
    plt.plot([x[1] for x in pop], label="Predator")
    plt.xlabel('Time')
    plt.ylabel('Population')
    plt.legend()
    plt.show()
```

```python
A = [.4, 0.2, .5]
B = [.3, 0.4, 0.4]
C = [1.5, 2, 1.3]
K = [10, 20, 15]
R = [2, 5, 1]

for i in range(3):
    prey_predator_model(A[i], B[i], C[i], K[i], R[i])
```
